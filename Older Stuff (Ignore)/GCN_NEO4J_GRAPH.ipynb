```python
from neo4j import GraphDatabase

URI = "bolt://localhost:7687"
AUTH = ("neo4j", "password")

driver = GraphDatabase.driver(URI, auth=AUTH)

def create_ip_node(ip_address, ip_type):
    def add_ip(tx, ip_address, ip_type):
        query = (
            "MERGE (ip:IP {address: $ip_address, type: $ip_type})"
        )
        tx.run(query, ip_address=ip_address, ip_type=ip_type)

    with driver.session() as session:
        session.write_transaction(add_ip, ip_address, ip_type)

def create_port_node(port_number, port_type):
    def add_port(tx, port_number, port_type):
        query = (
            "MERGE (port:Port {number: $port_number, type: $port_type})"
        )
        tx.run(query, port_number=port_number, port_type=port_type)

    with driver.session() as session:
        session.write_transaction(add_port, port_number, port_type)

def create_relationship(src_ip, src_port, dst_ip, dst_port, protocol):
    def create_connection(tx, src_ip, src_port, dst_ip, dst_port, protocol):
        query = (
            "MATCH (src:IP {address: $src_ip}), (dst:IP {address: $dst_ip}), "
            "(sport:Port {number: $src_port}), (dport:Port {number: $dst_port}) "
            "MERGE (src)-[:USES]->(sport) "
            "MERGE (src)-[:CONNECTS_TO {protocol: $protocol}]->(dst) "
            "MERGE (dst)-[:LISTENS_ON]->(dport)"
        )
        tx.run(query, src_ip=src_ip, src_port=src_port, dst_ip=dst_ip, dst_port=dst_port, protocol=protocol)

    with driver.session() as session:
        session.write_transaction(create_connection, src_ip, src_port, dst_ip, dst_port, protocol)
```

```python
# Example usage with your dataset
# Assume you have a DataFrame 'df' with columns 'srcip', 'sport', 'dstip', 'dsport', 'proto'

for index, row in df.iterrows():
    # Create nodes for source and destination IPs and ports
    create_ip_node(row['srcip'], "source")
    create_ip_node(row['dstip'], "destination")
    create_port_node(row['sport'], "source")
    create_port_node(row['dsport'], "destination")

    # Create relationships between these nodes
    create_relationship(row['srcip'], row['sport'], row['dstip'], row['dsport'], row['proto'])
```

```python
from neo4j import GraphDatabase
import torch
from torch_geometric.data import Data

URI = "bolt://localhost:7687"
AUTH = ("neo4j", "password")
driver = GraphDatabase.driver(URI, auth=AUTH)

def map_ip_to_idx(ip_address, ip_to_idx_map):
    if ip_address not in ip_to_idx_map:
        # Assign a new index to the IP address if it's not in the map
        ip_to_idx_map[ip_address] = len(ip_to_idx_map)
    return ip_to_idx_map[ip_address]

# You would also need a function to calculate the number of unique nodes
def calculate_num_nodes(ip_to_idx_map):
    return len(ip_to_idx_map)

# Initialize a map to keep track of IP addresses and their assigned indices
ip_to_idx_map = {}

def fetch_graph_data():
    query = """
    MATCH (src:IP)-[c:CONNECTS_TO]->(dst:IP)
    RETURN src.address AS source, dst.address AS target, c.protocol AS protocol
    """
    with driver.session() as session:
        result = session.run(query)
        edges = []
        for record in result:
            src_idx = map_ip_to_idx(record['source'], ip_to_idx_map)
            dst_idx = map_ip_to_idx(record['target'], ip_to_idx_map)
            edges.append([src_idx, dst_idx])
            # Add features as needed
        return torch.tensor(edges).t().contiguous()

edges = fetch_graph_data()
feature_dim=4

# Assuming all nodes have the same feature dimension
num_nodes = calculate_num_nodes(ip_to_idx_map)
node_features = torch.randn(num_nodes, feature_dim)  # Random features for illustration

data = Data(x=node_features, edge_index=edges)
```

```python
import torch
from torch.nn import Linear
from torch_geometric.nn import GCNConv

class GCN(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super().__init__()
        torch.manual_seed(1234)
        self.conv1 = GCNConv(num_features, 32)
        self.conv2 = GCNConv(32, 4)
        self.conv3 = GCNConv(4, 2)
        self.classifier = Linear(2, num_classes)

    def forward(self, x, edge_index):
        h = self.conv1(x, edge_index)
        h = h.tanh()
        h = self.conv2(h, edge_index)
        h = h.tanh()
        h = self.conv3(h, edge_index)
        h = h.tanh()  # Final GNN embedding space.
        
        # Apply a final (linear) classifier.
        out = self.classifier(h)

        return out

num_features = data.num_features
num_classes = 2  
model = GCN(num_features, num_classes)
print(model)
```

```python
import torch
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, data):
    model.eval()  

    with torch.no_grad():
        out = model(data.x, data.edge_index)

    _, predicted = torch.max(out, 1)

    test_labels = data.y[data.test_mask]
    test_preds = predicted[data.test_mask]

    # Calculate evaluation metrics
    accuracy = accuracy_score(test_labels.cpu(), test_preds.cpu())
    precision = precision_score(test_labels.cpu(), test_preds.cpu(), average='weighted')
    recall = recall_score(test_labels.cpu(), test_preds.cpu(), average='weighted')
    f1 = f1_score(test_labels.cpu(), test_preds.cpu(), average='weighted')

    return accuracy, precision, recall, f1
accuracy, precision, recall, f1 = evaluate_model(model, data)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
import torch
from torch_geometric.data import Data
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your PyTorch Geometric Data object
# Assuming 'model_class' is your model class, e.g., GCN

num_nodes = data.num_nodes
k = 10  # Number of folds
kf = KFold(n_splits=k, shuffle=True, random_state=42)

def train_model(train_index, test_index):
    # Initialize model
    model = GCN(num_features, num_classes)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

    # Create train and test masks
    train_mask = torch.zeros(num_nodes, dtype=torch.bool)
    test_mask = torch.zeros(num_nodes, dtype=torch.bool)
    train_mask[train_index] = True
    test_mask[test_index] = True

    # Training loop
    for epoch in range(100):
        model.train()
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = F.nll_loss(out[train_mask], data.y[train_mask])
        loss.backward()
        optimizer.step()

    # Evaluation
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
    y_pred = out.argmax(dim=1)
    accuracy = accuracy_score(data.y[test_mask].cpu(), y_pred[test_mask].cpu())
    return accuracy, y_pred[test_mask].cpu()

# Cross-validation
accuracies = []
for train_index, test_index in kf.split(range(num_nodes)):
    accuracy, y_pred = train_model(train_index, test_index)
    accuracies.append(accuracy)

    # Print classification report and confusion matrix for each fold
    print(classification_report(data.y[test_index], y_pred))
    plt.figure(figsize=(5,5))
    sns.heatmap(confusion_matrix(data.y[test_index], y_pred), annot=True, cmap='GnBu', fmt='d')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Overall performance
print('Average Accuracy:', sum(accuracies) / k)
```
